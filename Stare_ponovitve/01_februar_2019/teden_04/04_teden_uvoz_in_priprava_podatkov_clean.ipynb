```python
def remove_entry(record):
    try:
        del record['location']
    except KeyError:
        pass
    
    return record
    
```

```python
def parse_json(record):
    return json.loads(record, object_hook=remove_entry)
```

### Cleaning the Entire Dataset Using the applymap Function

```python
!head -n 20 data/DATA_university_towns.txt
```

```python
university_towns = []
```

```python
with open('data/DATA_university_towns.txt') as file:
    for line in file:
        if '[edit]' in line:
            state = line
        else:
            university_towns.append((state, line))
```

```python
university_towns[:18]
```

```python
towns_df = pd.DataFrame(university_towns, columns=['State', 'RegionName'])
```

```python
towns_df.head(15)
```

```python
def clean_df(item):
    if ' (' in item:
        return item[:item.find(' (')]
    elif '[' in item:
        return item[:item.find('[')]
    else:
        return item
```

```python
towns_df = towns_df.applymap(clean_df)
```

```python
towns_df.head(15)
```
