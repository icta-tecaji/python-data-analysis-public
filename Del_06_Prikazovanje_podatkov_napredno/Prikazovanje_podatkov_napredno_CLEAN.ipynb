### Hiding Spines

```python
fig, ax = plt.subplots()

ax.plot(women_degrees["Year"], women_degrees["Biology"], label="Women")
ax.plot(women_degrees["Year"], 100-women_degrees["Biology"], label="Men")
ax.legend()
ax.set_title("Percentage of Biology Degrees by Gender")

ax.tick_params(left=False, bottom=False)

for key, spine in ax.spines.items():
    spine.set_visible(False)

plt.show()
```

### Comparing data with multiple plots

<p><img alt="Four Major Categories" src="images/four_major_categories_plots.png"></p>

```python
major_cats = ["Biology", "Computer Science", "Engineering", "Math and Statistics"]
fig = plt.figure(figsize=(12, 12))

for sp in range(0, 4):
    ax = fig.add_subplot(2, 2, sp+1)
    ax.plot(women_degrees["Year"], women_degrees[major_cats[sp]], label="Women")
    ax.plot(women_degrees["Year"], 100-women_degrees[major_cats[sp]], label="Men")
    ax.set_title(major_cats[sp])
    ax.tick_params(left=False, bottom=False)
    ax.set_ylim(0, 100)
    for key, spine in ax.spines.items():
        spine.set_visible(False)

        
ax.legend()
plt.show()
```

## Color, Layout, and Annotations

### Setting Line Color Using RGB

<p><img alt="Tableau RGB values" src="images/tableau_rgb_values.png"></p>

```python
cb_dark_blue = (200/255, 82/255, 0/255)
cb_orange = (255/255, 188/255, 121/255)

major_cats = ["Biology", "Computer Science", "Engineering", "Math and Statistics"]
fig = plt.figure(figsize=(12, 12))

for sp in range(0, 4):
    ax = fig.add_subplot(2, 2, sp+1)
    ax.plot(women_degrees["Year"], women_degrees[major_cats[sp]], label="Women", c=cb_dark_blue)
    ax.plot(women_degrees["Year"], 100-women_degrees[major_cats[sp]], label="Men", c=cb_orange)
    ax.set_title(major_cats[sp])
    ax.tick_params(left=False, bottom=False)
    ax.set_ylim(0, 100)
    for key, spine in ax.spines.items():
        spine.set_visible(False)

        
ax.legend()
plt.show()
```

### Setting Line Width

```python
cb_dark_blue = (200/255, 82/255, 0/255)
cb_orange = (255/255, 188/255, 121/255)

major_cats = ["Biology", "Computer Science", "Engineering", "Math and Statistics"]
fig = plt.figure(figsize=(12, 12))

for sp in range(0, 4):
    ax = fig.add_subplot(2, 2, sp+1)
    ax.plot(women_degrees["Year"], women_degrees[major_cats[sp]], label="Women", c=cb_dark_blue, linewidth=3)
    ax.plot(women_degrees["Year"], 100-women_degrees[major_cats[sp]], label="Men", c=cb_orange, linewidth=3)
    ax.set_title(major_cats[sp])
    ax.tick_params(left=False, bottom=False)
    ax.set_ylim(0, 100)
    for key, spine in ax.spines.items():
        spine.set_visible(False)

        
ax.legend()
plt.show()
```

### Replacing the Legend With Annotations

<p><img alt="Annotated Legend" src="images/annotated_legend.png"></p>

```python
major_cats = ["Engineering", "Computer Science", "Psychology", "Biology", "Physical Sciences", "Math and Statistics"]

fig = plt.figure(figsize=(20, 4))

for sp in range(0, len(major_cats)):
    ax = fig.add_subplot(1, len(major_cats), sp+1)
    ax.plot(women_degrees["Year"], women_degrees[major_cats[sp]], linewidth=3)
    ax.plot(women_degrees["Year"], 100-women_degrees[major_cats[sp]], linewidth=3)
    ax.tick_params(left=False, bottom=False)
    ax.set_ylim(0, 100)
    for key, spine in ax.spines.items():
        spine.set_visible(False)
    ax.set_title(major_cats[sp])
    
    if sp == 0:
        ax.text(2005, 87, "Men")
        ax.text(2005, 8, "Women")
        
    elif sp == len(major_cats)-1:
        ax.text(2005, 62, "Men")
        ax.text(2005, 35, "WOmen")
        
plt.show()
```
